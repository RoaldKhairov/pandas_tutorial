## Map

<center> <img src="imgs/1200px-Bataille_entre_mongols_&_chinois_(1211).jpeg" width=600></img>

```python
df['platform'].unique()
```

```python
mask = {
    'android': 1, 
    'desktop': 0, 
    'ios': 1, 
    'mobile': 1, 
    'unknown': 0
}
df['platform_coded'] = df['platform'].map(mask)
df.head()
```

### GroupBy

<center> <img src="imgs/hussars.jpg" width=600></img>

<center> <img src="imgs/groupby.jpg" width=600></img>

```python
df.groupby('region')['platform_coded']
```

```python
df.groupby('region')['platform_coded'].mean().head()
```

```python
columns_to_show = ['platform_coded', 'cat_id']

table = df.groupby(['region'])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
table.head()
```

```python
df.groupby('region')['cat_id'].agg(
    ['mean', 'std', 'size', lambda x: max(x)]).head()
```
